## Anthropic's agent demo

Let's replicate Anthropic's [Customer Service Agent demo](https://github.com/anthropics/claude-cookbooks/blob/main/tool_use/customer_service_agent.ipynb) using Solveit.

````
orders = {
    "O1": dict(id="O1", product="Widget A", quantity=2, price=19.99, status="Shipped"),
    "O2": dict(id="O2", product="Gadget B", quantity=1, price=49.99, status="Processing"),
    "O3": dict(id="O3", product="Gadget B", quantity=2, price=49.99, status="Shipped")}

customers = {
    "C1": dict(name="John Doe", email="john@example.com", phone="123-456-7890",
                orders=[orders['O1'], orders['O2']]),
    "C2": dict(name="Jane Smith", email="jane@example.com", phone="987-654-3210",
                orders=[orders['O3']])
}
````

````
def get_customer_info(
    customer_id:str # ID of the customer
): # Customer's name, email, phone number, and list of orders
    "Retrieves a customer's information and their orders based on the customer ID"
    return customers.get(customer_id, "Customer not found")

def get_order_details(
    order_id:str # ID of the order
): # Order's ID, product name, quantity, price, and order status
    "Retrieves the details of a specific order based on the order ID"
    return orders.get(order_id, "Order not found")

def cancel_order(
    order_id:str # ID of the order to cancel
)->bool: # True if the cancellation is successful
    "Cancels an order based on the provided order ID"
    if order_id not in orders: return False
    orders[order_id]['status'] = 'Cancelled'
    return True
````

Here's the complete solution:

```python
from aocd.models import Puzzle

# Get the puzzle data
puzzle = Puzzle(year=2024, day=1)
input_data = puzzle.input_data

# Parse the input into two lists
left = []
right = []
for line in input_data.strip().split('\n'):
    l, r = map(int, line.split())
    left.append(l)
    right.append(r)

# Sort both lists
left_sorted = sorted(left)
right_sorted = sorted(right)

# Calculate sum of absolute differences
answer = sum(abs(l - r) for l, r in zip(left_sorted, right_sorted))
print(f"Part 1 answer: {answer}")
```

**How it works:**

1. **Parse the input**: Each line contains two numbers separated by spaces. We split each line and convert the numbers to integers, building two separate lists.

2. **Sort both lists**: We sort each list independently in ascending order.

3. **Pair and calculate differences**: Using `zip()`, we pair up the first element from each sorted list, then the second elements, and so on. For each pair, we calculate the absolute difference.

4. **Sum the differences**: We add up all the differences to get the final answer.

**Example walkthrough:**
- Input: `[3,4,2,1,3,3]` and `[4,3,5,3,9,3]`
- After sorting: `[1,2,3,3,3,4]` and `[3,3,3,4,5,9]`
- Differences: `|1-3|=2, |2-3|=1, |3-3|=0, |3-4|=1, |3-5|=2, |4-9|=5`
- Sum: `2+1+0+1+2+5 = 11`
